# **Theory Questions**
### Q1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating a null hypothesis (\(H_0\)) and an alternative hypothesis (\(H_1\)), then using sample data to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

### Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis (\(H_0\)) is a statement that there is no effect or no difference, and it serves as the default assumption in hypothesis testing. The alternative hypothesis (\(H_1\)) is a statement that indicates the presence of an effect or a difference. The key difference is that the null hypothesis assumes no change, while the alternative hypothesis suggests that a change or effect exists.

### Q3. What is the significance level in hypothesis testing, and why is it important?
The significance level (\(\alpha\)) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It is important because it defines the threshold for determining whether the observed data is statistically significant. Common significance levels are 0.05, 0.01, and 0.10.

### Q4. What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming that the null hypothesis is true. It quantifies the evidence against the null hypothesis.

### Q5. How do you interpret the P-value in hypothesis testing?
A low P-value (typically less than the significance level \(\alpha\)) indicates strong evidence against the null hypothesis, leading to its rejection. Conversely, a high P-value suggests insufficient evidence to reject the null hypothesis.

### Q6. What are Type 1 and Type 2 errors in hypothesis testing?
- **Type I error** occurs when the null hypothesis is rejected when it is actually true (false positive).
- **Type II error** occurs when the null hypothesis is not rejected when it is actually false (false negative).

### Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
A one-tailed test evaluates the possibility of the relationship in one direction (either greater than or less than), while a two-tailed test evaluates the possibility in both directions (both greater than and less than).

### Q8. What is the Z-test, and when is it used in hypothesis testing?
The Z-test is a statistical test used to determine if there is a significant difference between sample and population means or between two sample means when the population variance is known or the sample size is large (typically \(n > 30\)).

### Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
The Z-score is calculated using the formula:

\[
Z = \frac{(X - \mu)}{\sigma}
\]

where \(X\) is the sample mean, \(\mu\) is the population mean, and \(\sigma\) is the population standard deviation. The Z-score represents the number of standard deviations a data point is from the mean.

### Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is a probability distribution used when the sample size is small (\(n < 30\)) and/or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, which accounts for the increased variability in smaller samples.

### Q11. What is the difference between a Z-test and a T-test?
The main difference is that a Z-test is used when the population variance is known or the sample size is large, while a T-test is used when the population variance is unknown and the sample size is small. The T-test also uses the T-distribution, which has heavier tails.

### Q12. What is the T-test, and how is it used in hypothesis testing?
The T-test is a statistical test used to compare the means of two groups or to compare a sample mean to a known population mean when the population variance is unknown. It helps determine if the differences between groups are statistically significant.

### Q13. What is the relationship between Z-test and T-test in hypothesis testing?
Both the Z-test and T-test are used to test hypotheses about means. The Z-test is appropriate for large samples or known population variances, while the T-test is used for small samples or unknown population variances. As the sample size increases, the T-distribution approaches the normal distribution, making the Z-test and T-test results converge.

### Q14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval is a range of values, derived from sample statistics, that is likely to contain the population parameter with a specified level of confidence (e.g., 95%). It provides an estimate of the uncertainty around the sample mean and helps in understanding the precision of the estimate.

### Q15. What is the margin of error, and how does it affect the confidence interval?
The margin of error is the amount of error allowed in the estimate of a population parameter. It is half the width of the confidence interval. A larger margin of error results in a wider confidence interval, indicating less precision, while a smaller margin of error results in a narrower confidence interval, indicating more precision.

### Q16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem is a mathematical formula used to update the probability of a hypothesis based on new evidence. It is significant because it allows for the incorporation of prior knowledge or beliefs into statistical analysis, providing a more comprehensive understanding of uncertainty.

### Q17. What is the Chi-square distribution, and when is it used?
The Chi-square distribution is a probability distribution used in hypothesis testing, particularly for categorical data. It is commonly used in tests of independence and goodness of fit, where it assesses how well observed data fit expected data.

### Q18. What is the Chi-square goodness of fit test, and how is it applied?
The Chi-square goodness of fit test is used to determine if a sample distribution matches an expected distribution. It compares the observed frequencies of categories to the expected frequencies and calculates a Chi-square statistic to assess the fit.

### Q19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a probability distribution used primarily in analysis of variance (ANOVA) and regression analysis. It is used to compare variances between two or more groups to determine if they are significantly different.

### Q20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups. Its assumptions include:
1. Independence of observations.
2. Normally distributed populations.
3. Homogeneity of variances (equal variances across groups).

### Q21. What are the different types of ANOVA tests?
The main types of ANOVA tests include:
1. One-way ANOVA: Compares means across one independent variable.
2. Two-way ANOVA: Compares means across two independent variables and their interaction.
3. Repeated measures ANOVA: Compares means across multiple measurements from the same subjects.

### Q22. What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test used to compare the variances of two or more groups. It is often used in the context of ANOVA to determine if there are significant differences between group means based on their variances. The F-statistic is calculated as the ratio of the variance between groups to the variance within groups.

```python
# Q1: Perform a Z-test for comparing a sample mean to a known population mean
def z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    n = len(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    return z_statistic, p_value
# Simulate sample data
np.random.seed(0)
sample_data = np.random.normal(loc=5, scale=1, size=30)
population_mean = 5.5
population_std = 1
z_stat, p_val = z_test(sample_data, population_mean, population_std)
print("Z-Statistic:", z_stat)
print("P-Value:", p_val)
```

```python
# Q2: Simulate random data to perform hypothesis testing and calculate the corresponding P-value
random_sample = np.random.normal(loc=10, scale=2, size=50)
population_mean_test = 9
z_stat_test, p_val_test = z_test(random_sample, population_mean_test, 2)
print("Z-Statistic for random sample:", z_stat_test)
print("P-Value for random sample:", p_val_test)
```

```python
# Q3: Implement a one-sample Z-test
def one_sample_z_test(sample, population_mean, population_std):
    return z_test(sample, population_mean, population_std)
```

```python
# Q4: Perform a two-tailed Z-test and visualize the decision region
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)
plt.figure(figsize=(10, 5))
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x), label='Standard Normal Distribution')
plt.fill_between(x, stats.norm.pdf(x), where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.axvline(z_critical, color='red', linestyle='--')
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Q5: Calculate and visualize Type 1 and Type 2 errors
def visualize_errors(sample_size, population_mean, population_std, alpha):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    z_stat, p_val = z_test(sample, population_mean, population_std)

    # Type I Error
    z_critical = stats.norm.ppf(1 - alpha/2)
    type_1_error = p_val < alpha

    # Type II Error
    alternative_mean = population_mean + 1  # Shift the mean for Type II error
    z_stat_alt = (np.mean(sample) - alternative_mean) / (population_std / np.sqrt(sample_size))
    p_val_alt = 2 * (1 - stats.norm.cdf(abs(z_stat_alt)))
    type_2_error = p_val_alt > alpha

    print("Type I Error:", type_1_error)
    print("Type II Error:", type_2_error)
visualize_errors(30, 5, 1, 0.05)
```

```python
# Q9: Calculate the confidence interval for a sample mean
def confidence_interval(sample, confidence_level):
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    z_critical = stats.norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z_critical * std_err
    return (mean - margin_of_error, mean + margin_of_error)
ci = confidence_interval(sample_data, 0.95)
print("Confidence Interval:", ci)
```

```python
# Q10: Calculate the margin of error for a given confidence level using sample data
def margin_of_error(sample, confidence_level):
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    z_critical = stats.norm.ppf((1 + confidence_level) / 2)
    return z_critical * std_err
me = margin_of_error(sample_data, 0.95)
print("Margin of Error:", me)
```
